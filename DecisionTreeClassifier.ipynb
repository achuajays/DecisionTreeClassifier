```python
import pandas as pd
import numpy as np
```

```python
n = pd.read_csv('/content/drug200.csv')
```

```python
n.head()
```

```python
n.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import r2_score
```

```python


n['BP'].replace({'NORMAL': 0, 'HIGH': 1, 'LOW': 2}, inplace=True)
n['Cholesterol'].replace({'NORMAL': 0, 'HIGH': 1, 'LOW': 2}, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Example categorical data


# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the data
n['Sex'] = label_encoder.fit_transform(n['Sex'])
n['drug'] = label_encoder.fit_transform(n['Drug'])
d = list(zip(n['Drug'], n['drug']))
def value_d():
  for i ,j in d:
    print(i , j)
```

```python
y = n['drug']
X = n[['Age',	'Sex',	'BP',	'Cholesterol',	'Na_to_K'	]]
```

```python
X
```

```python
x_t , x_te , y_t , y_te = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
x_t
```

```python
x_te
```

```python
y_t
```

```python
y_te
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x_t,y_t)
```

```python
p = model.predict(x_te)
```

```python
p
```

```python
r2_score(p,y_te)
```
